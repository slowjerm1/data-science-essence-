```python
'''Ways to Cleanse Missing Data in Python
To perform a Python data cleansing,
you can drop the missing values, replace them, 
replace each NaN with a scalar value, or fill forward or backward.
'''
```

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r'C:\Users\slowjerm.jerm1\worldometer_data.csv')
data
```

```python
data.isnull().any()
```

```python
#View total of null values by column
data.isnull().sum()
```

```python
#View the number of null values in the 'TotalRecovered ' column
data['TotalRecovered'].isnull().sum()
```

```python
#Fill all null values with a space, and score that in the current data frame
data=data.fillna(" ")
data
```

```python
Store the dataframe as a new CSV
data.to_csv('newCovid_Cases.csv', index=False)
```

```python
data.dropna()
```

```python
#Replacing Missing Values
from numpy import NaN

data.replace({NaN:0.00})
```

```python
'''Replacing with a Scalar Value
We can use the fillna() method for this.'''

data.fillna(7)
```

```python
'''Filling Forward or Backward
If we supply a method parameter to the fillna() method, we can fill forward or backward as we need. 
To fill forward, use the methods pad or fill, and to fill backward, use bfill and backfill.'''

data.fillna(method='pad')
```

```python
'''Filling Forward or Backward
If we supply a method parameter to the fillna() method, we can fill forward or backward as we need. To fill forward,
use the methods pad or fill, and to fill backward, use bfill and backfill.'''

data.fillna(method='backfill')
```
